```python
from PIL import Image
import glob
import os
import leargist
from scipy import spatial
from operator import itemgetter
import cPickle as pickle

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

## Display the images

```python
InteractiveShell.ast_node_interactivity = "none"
import matplotlib.pyplot as plt
%matplotlib inline 

im1 = Image.open('../data/example_mobile_image.jpg')
im2 = Image.open('../data/example_segmented_mobile_image.jpg')
im3 = Image.open('../data/example_catalogue_image.jpg')
im4 = Image.open('../data/example_catalogue_image2.jpg')

plt.figure(figsize=(15,10))

plt.subplot(1,4,1)
plt.imshow(im1)
plt.title('mobile')

plt.subplot(1,4,2)
plt.imshow(im2)
plt.title('segmented mobile')

plt.subplot(1,4,3)
plt.imshow(im3)
plt.title('catalouge')

plt.subplot(1,4,4)
plt.imshow(im4)
plt.title('catalouge')
```

## Calculagte GIST

```python
InteractiveShell.ast_node_interactivity = "all"

gist1 =  leargist.color_gist(im1)
gist2 =  leargist.color_gist(im2)
gist3 =  leargist.color_gist(im3)
gist4 =  leargist.color_gist(im4)

gist1.shape
gist1
```

## Get the distance between GIST descriptors of segmented mobile image and catalogue image

```python
# distance between mobile and catalogue
distance13 = spatial.distance.euclidean(gist1, gist3)
'Distance between GIST descriptors of mobile and catalouge image: {0}'.format(distance13)

# distance between segmented mobile and catalogue
distance23 = spatial.distance.euclidean(gist2, gist3)
'Distance between GIST descriptors of segmented mobile and catalouge image: {0}'.format(distance23)

# distance between segmented mobile and different catalogue
distance24 = spatial.distance.euclidean(gist2, gist4)
'Distance between GIST descriptors of segmented mobile and different catalouge image: {0}'.format(distance24)
```

That means that **segmented mobile** image is more closely matched to **its own catalogue** image.
